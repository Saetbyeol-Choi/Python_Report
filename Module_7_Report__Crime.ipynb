## **3) One column selection**

Given that different cities have different populations, we sorted the data in highest order based on the 'crimes_percapita' column. This shows that Miami, located in Florida had the highest number of crime rate per population in 1990 at 4352.83. The city had the first and second highest crime rate by population of the top 10 data. Therefore, based on these results, we would like to provide additional information on the crime rate per capita by crime type in the city, with Miami as a priority.

```python
# The "dropna" function finds rows with NaN and deletes the entire row.
report2 = report.dropna()
```

```python
# 10 rows have the largest values on the 'crimes_percapita' column are extracted. 
report2.nlargest(10, "crimes_percapita")
```

## **4) Sort**

Based on the results derived above, we provide information by analyzing the ratio of crime types per population in Miami, Florida. There are four types of crime: homicide, rape, assault and robbery. By listing each type of crime in order of highest record, it can be determined that which year had the highest crime rate per capita. By extracting the top 5 values for each, the crime rate can be seen by year and compared.<br>All four types of crime were identified as having high records throughout the 1970s and 1990s. Among the four types of crime, the crimes with the overwhelmingly highest number of crimes were robbery and assault.

```python
# Among the 'agency_jurisdiction' column, the rows corresponding to Miami,FL are extracted.

miami_fl = report2.loc[report2["agency_jurisdiction"] == "Miami, FL" ]
```

```python
# Create a new dataset with 5 columns.

mia_report = miami_fl[["report_year", "homicides_percapita", "rapes_percapita", "assaults_percapita", "robberies_percapita"]]
mia_report.head()
```

```python
# Sort the data in descending order by homicides per capita. The "head" function extracts the rows corresponding to the top 5 values. 

homicides = mia_report.sort_values(by=["homicides_percapita"], ascending = [False])
homicides.head()
```

```python
# Same as the above, the top 5 values are shown after sorting the rape rate per population in descending order.

rapes = mia_report.sort_values(by=["rapes_percapita"], ascending = [False])
rapes.head()
```

```python
# The data on the assault rate per capita are also shown in the same way as above.

assaults = mia_report.sort_values(by=["assaults_percapita"], ascending = [False])
assaults.head()
```

```python
# The data on the robbery rate per capita are also shown in the same way as above.

robberies = mia_report.sort_values(by=["robberies_percapita"], ascending = [False])
robberies.head()
```
